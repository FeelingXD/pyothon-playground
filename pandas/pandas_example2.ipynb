```python
# 데이터는 다음과같다.

"""
    pyupbit : 파이썬 업비트데이터를 사용합니다.
    open : 시작가
    high : 고가
    low : 저가
    closevolume : 종가         
    value  : 가치 
"""
df.info()

df.head()
# 단일열 선택
df['open']

# 여러열 선택
df[['high','value']] #  이차원 배열 로 파라미터를 넘겨준다.
```

```python
# iterrow로 행순회하기

for index, row in df.iterrows():
    if row.get('high'):
        print(row['high'])
# df.high=df['high']
# df.low=df['low']
df=df.assign(is_increase=lambda x : x['open']<x['close'],)

# df.loc[df.high>10000&df.low>50000]

# close shift 같은행 이전데이터의 연산이 필요할경우

df['전일종가']= df.close.shift(1)
df
```
