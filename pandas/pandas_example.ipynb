```python
import pandas as pd


# 판다스 데이터 만들기 일반 이중 배열일경우
test_data1 = [['a',1],['b',2]]
# 이중배열 이 데이터가 되고 columns로 인덱스를 잡는다.
frame1 = pd.DataFrame(test_data1,columns=['first','second'])
print(frame1)

# 딕셔너리로 데이터프레임을 만드는방법
# 2차원 표로서의 데이터를 표현
# key 가 칼럼명이되고 values 가 values 가 됨
test_data2 ={"종가" :[1,2,3,4,5] ,"PER" :[30,40,50,60,70],"PBR":[1,2,3,4,5]}
frame2 = pd.DataFrame(test_data2)
print(frame2)

## 인덱스를 만들어주기
# 인덱스를 설정하지않으면 기본적으로 0~ AI 값이 설정된다.
INDEX =['NAVER','KAKAO','한전','SAMSUNG','SK']
frame2=pd.DataFrame(test_data2,INDEX)
frame2
```

```python
## 판다스에서 칼럼을 선택하는방법 

## 위의에졔 파일을 추가적으로 사용함 

frame2
## 내부배열로 칼럼의 명칭을 사용하면 해당 데이터만볼수 있음
frame2[['종가']] 

## 배열에 값을 추가함으로써 멀티칼럼도지원함
frame2[['종가','PER']]
```

```python
## 로우를 선택하는방법

## 데이터를 다룰때 실제 로우(행)을 다룰때의 방법을 알아본다.

# iloc 인덱스번호 를 사용해서 row를 선택

frame2.iloc[1]
# loc 을사용해서 인덱스 이름을 사용해서 row를 선택
frame2.loc['KAKAO']


# 행 슬라이스 기법
# 파이썬에서의 슬라이스처럼 판다스에서도 슬라이스를 지원한다.
frame2

# iloc 을 사용한 인덱스 번호기반 슬라이싱
frame2.iloc[1:3]

# loc 을 사용한 인덱스 이름기반 슬라이싱
frame2.loc['KAKAO':'한전']

## drop 해보기 
frame2.drop(['종가'], axis=1)

frame2['v']=frame2['PBR']+frame2['PER']

frame2
```
